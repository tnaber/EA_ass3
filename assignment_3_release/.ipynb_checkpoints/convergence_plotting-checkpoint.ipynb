```python
import subprocess
import sys
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
domain = [200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200, 1300, 1400, 1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200, 2300, 2400, 2500, 2600, 2700, 2800, 2900, 3000, 3100, 3200, 3300, 3400, 3500, 3600, 3700, 3800, 3900, 4000, 4100, 4200, 4300, 4400, 4500, 4600, 4700, 4800, 4900, 5000, 5100, 5200, 5300, 5400, 5500, 5600, 5700, 5800, 5900, 6000, 6100, 6200, 6300, 6400, 6500, 6600, 6700, 6800, 6900, 7000, 7100, 7200, 7300, 7400, 7500, 7600, 7700, 7800, 7900, 8000, 8100, 8200, 8300, 8400, 8500, 8600, 8700, 8800, 8900, 9000, 9100, 9200, 9300, 9400, 9500, 9600, 9700, 9800, 9900, 10000]

data = np.load("data/hypervolume_results_L_20pop_100.npy")
labels = ["With elitism", "Without elitism"]
```

```python

for i in range(len(data)):
	res = data[i]
	interdecile_range_low = np.percentile(res, 10, axis=0)
	interdecile_range_high = np.percentile(res, 90, axis=0)

	average = np.average(res, axis=0)

	plt.plot(domain, average)
	plt.fill_between(domain,
		interdecile_range_low,
		interdecile_range_high,
					alpha=0.3)
    
plt.xlabel('Amount of evaluations')
plt.ylabel('Hypervolume')
plt.legend(labels) # [("l = " + x) for x in L]
fig = plt.gcf()
plt.show()
```
