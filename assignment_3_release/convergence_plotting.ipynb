```python
# print(data)
for i in range(len(data)):
	domain = data[i][0][0]
	res = [np.array(x) for x in data[i][:,1]]

	interdecile_range_low = np.percentile(res, 10, axis=0)
	interdecile_range_high = np.percentile(res, 90, axis=0)

	average = np.average(res, axis=0)

	plt.plot(domain, average)
	plt.fill_between(domain,
		interdecile_range_low,
		interdecile_range_high,
					alpha=0.3)
    
plt.xlabel('Amount of evaluations')
plt.ylabel('Hypervolume')
# plt.yscale('log')
plt.legend(labels, loc='lower right') # [("l = " + x) for x in L]
fig = plt.gcf()
plt.show()
```

```python
fig.tight_layout(pad=3.0)
fig.savefig('plots/' + file_name)
```
